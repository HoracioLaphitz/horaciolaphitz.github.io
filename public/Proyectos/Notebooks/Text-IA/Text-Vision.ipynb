```python
!pip install tensorflow-text
import tensorflow as tf
import tensorflow_hub as hub
import tensorflow_text as text
```

```python
# definir un modelo de incrustación de texto
text_input = tf.keras.layers.Input(shape=(), dtype=tf.string)
preprocessor = hub.KerasLayer("https://kaggle.com/models/jeongukjae/klue-roberta/frameworks/TensorFlow2/variations/cased-preprocess/versions/1")
encoder_inputs = preprocessor(text_input)

encoder = hub.KerasLayer("https://www.kaggle.com/models/jeongukjae/klue-roberta/frameworks/TensorFlow2/variations/cased-l-12-h-768-a-12/versions/1", trainable=True)
encoder_outputs = encoder(encoder_inputs)
pooled_output = encoder_outputs["pooled_output"]      # [batch_size, 768].
sequence_output = encoder_outputs["sequence_output"]  # [batch_size, seq_length, 768].


# creacion del modelo
model = tf.keras.Model(text_input, pooled_output)

#agregar oraciones
sentences = tf.constant(["()"])
print(model(sentences))
```

```python
preprocessor = hub.load("https://kaggle.com/models/jeongukjae/klue-roberta/frameworks/TensorFlow2/variations/cased-preprocess/versions/1")
tokenize = hub.KerasLayer(preprocessor.tokenize)
bert_pack_inputs = hub.KerasLayer(preprocessor.bert_pack_inputs)
encoder = hub.KerasLayer("https://www.kaggle.com/models/jeongukjae/klue-roberta/frameworks/TensorFlow2/variations/cased-l-12-h-768-a-12/versions/1", trainable=True)

text_inputs = [
    tf.keras.layers.Input(shape=(), dtype=tf.string),
    tf.keras.layers.Input(shape=(), dtype=tf.string),
]
tokenized_inputs = [tokenize(segment) for segment in text_inputs]
encoder_inputs = bert_pack_inputs(tokenized_inputs)
encoder_outputs = encoder(encoder_inputs)

pooled_output = encoder_outputs["pooled_output"]      # [batch_size, 768].
sequence_output = encoder_outputs["sequence_output"]  # [batch_size, seq_length, 768].

model = tf.keras.Model(text_inputs, pooled_output)

# Predicciones
hypotheses = tf.constant(["(Hola mundo)"])
premises = tf.constant(["(Hola Mundo)"])

#print(model([hypotheses, premises] / 100))
print(model([hypotheses, premises] / tf.constant(100, dtype=tf.float32)))


```
