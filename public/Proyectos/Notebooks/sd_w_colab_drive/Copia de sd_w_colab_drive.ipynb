```python
#@markdown #Descargar modelo desde civitai
import re
import requests
import json
import base64

tomate = base64.b64decode(('=kWdiV2dtQ2c'[::-1]).encode('ascii')).decode('ascii')

model_url = "" #@param {type:"string"}
cdownload_path = "" #@param {type:"string"}
cdownload_path = cdownload_path.replace("sd-tomate", tomate)


if cdownload_path=="":
  cdownload_path= "/content"

if model_url=="":
  model_url="a"

match = re.search(r"modelVersionId=(\d+)", model_url)

if match:
    modelVersionId = match.group(1)
    !wget -P $cdownload_path https://civitai.com/api/download/models/{modelVersionId} --content-disposition
else:
    if re.search(r"\bapi\b", model_url):
      coincidencia = re.findall(r"/models/(\d+)", model_url)
      if coincidencia:
        !wget -P $cdownload_path https://civitai.com/api/download/models/{coincidencia[0]} --content-disposition
      else:
        print("Error: URL inválida")
    else:
      coincidencia = re.findall(r"/models/(\d+)", model_url)
      if coincidencia:
        url = 'https://civitai.com/api/v1/models/' + coincidencia[0]
        headers = {'Content-Type': 'application/json'}
        response = requests.get(url, headers=headers)
        data = response.json()
        try:
          modelVersionId = str(data['modelVersions'][0]['id'])
          !wget -P $cdownload_path https://civitai.com/api/download/models/{modelVersionId} --content-disposition
        except:
          print("Error: URL inválida")
      else:
        print("Error: URL inválida")
```
